```python
def remove_10_row(df, column_name):

  counts = df[column_name].value_counts()

  # Выбор городов, встречающихся менее 10 раз
  red_city = counts[counts < 29].index

  # Удаление строк, содержащих редкие города
  df_filtered = df[~df[column_name].isin(red_city)]

  return df_filtered
  
```
