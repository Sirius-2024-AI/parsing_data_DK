```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("cian_dclick_clear.csv")
```

```python
data
```

```python
from geopy.distance import geodesic as GD
dist_list = list()
```

```python
def get_distance(row):
    
    center_lat = row["center_lat"]
    center_lon = row["center_lon"]
    
    lat = row["latitude"]
    lon = row["longitude"]
    
    dist_list.append(GD((center_lat, center_lon), (lat, lon)).km)
```

```python
#data['distance_to_center'] = None
```

```python
for index, row in data.iterrows():
    get_distance(row)
```

```python

data['distance_to_center'] = pd.Series(dist_list)
```

```python
data #GD((44.894818, 37.316367), (44.929875, 37.321874)).km
```

```python
data["area_price"] = data['price'] / data['area']
```

```python
data
```

```python
df_about_city = data.groupby('city').agg(mean_price_city = ("price", "mean"), mean_area_city=('area', 'mean'), mean_distance=("distance_to_center", "mean"), mean_squared_price = ('area_price', 'mean'))
```

```python
df_about_count_of_rooms = data.groupby('total_rooms').agg(mean_price=("price", "mean"), mean_area_of_flat=("area", "mean"))
```

```python
df_about_count_of_rooms
```

```python
df_about_city.to_csv("df_about_city.csv")
```

```python
data = data.merge(df_about_city, on="city", how="left")
```

```python
data = data.merge(df_about_count_of_rooms, on="total_rooms", how="left")
```
